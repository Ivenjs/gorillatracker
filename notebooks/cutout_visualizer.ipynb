```python
import os
import cv2
import random
import gorillatracker.utils.cutout_helpers as cutout_helpers
import gorillatracker.utils.visualizer_helpers as visualizer_helpers

base_path = "/workspaces/gorillatracker/data/ground_truth"
full_images_path = f"{base_path}/cxl/full_images/"
faces_images_path = f"{base_path}/cxl/face_images/"
body_images_path = "/workspaces/gorillatracker/data/derived_data/cxl/yolov8n_gorillabody_ybyh495y/body_images"
faces_images = os.listdir(faces_images_path)
```

```python
from IPython.display import Image, display


image = random.choice(faces_images)
face_image = cv2.imread(f"{faces_images_path}/{image}")
body_image = cv2.imread(f"{body_images_path}/{image}")
full_image = cv2.imread(f"{full_images_path}/{image}")
bbox_face = cutout_helpers.get_cutout_bbox(full_image, face_image)
bbox_body = cutout_helpers.get_cutout_bbox(full_image, body_image)
# show raw image
display(Image(filename=f"{full_images_path}/{image}"))
visualizer_helpers.show_bbox(full_image, bbox_face)
visualizer_helpers.show_bbox(full_image, bbox_body)
```
