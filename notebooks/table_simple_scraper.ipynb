```python
import json
from collections import defaultdict
from typing import Any, Literal
from dataclasses import dataclass
import wandb
from wandb.apis.public.runs import Runs, Run


@dataclass
class RunSummary:
    run: Run
    loss_mode: Literal["ntxent/l2sp", "online/soft/l2sp", "offline/native"]

    @property
    def metrics(self) -> dict[str, float]:
        return {
            "knn5-cv-val": self.run.summary.get("cxl-all-val/val/embeddings/knn5_crossvideo_cos/accuracy_max", float("nan")),
            "knn5-cv-test": self.run.summary.get("cxl-all-test/val/embeddings/knn5_crossvideo_cos/accuracy_max", float("nan")),
            "knn5-cv-all": self.run.summary.get("cxl-all/val/embeddings/knn5_crossvideo_cos/accuracy_max", float("nan")),
            "knn5-kfold": self.run.summary.get("cxl-kfold/val/embeddings/knn5_crossvideo_cos/accuracy_max", float("nan")),
        }

    def grouping_key(self, name: str) -> str:
        return self.run.config[name]


def get_runs(project: str) -> list[RunSummary]:
    run_summaries: list[RunSummary] = []
    api = wandb.Api()
    runs: Runs = api.runs(f"gorillas/{project}")
    for run in runs:
        if "loss_mode" not in run.config:
            print("Skipping", run)
            continue
        run_summaries.append(RunSummary(run, run.config["loss_mode"]))
    return run_summaries


def in_group(run: RunSummary, grouping_keys: list[str], group: list[str]) -> bool:
    for key, value in zip(grouping_keys, group, strict=True):
        # print(run.grouping_key(key), value)
        if run.grouping_key(key) != value:
            return False
    return True


def get_line(runs: list[RunSummary], grouping_keys: list[str], group: list[str]) -> RunSummary:
    for run in runs:
        if in_group(run, grouping_keys, group) and run.loss_mode == "ntxent/l2sp":
            return run
    raise ValueError(f"No SimCLR run found for group {group}")


def get_groups(runs: list[RunSummary], grouping: dict[str, Any]) -> defaultdict[str, list[tuple[str, RunSummary]]]:
    grouping_keys = grouping["group_by"]
    lines: defaultdict[str, list[tuple[str, RunSummary]]] = defaultdict(list)
    for group, groups in grouping["groups"].items():
        for g in groups:
            assert isinstance(g["group"], list)
            assert isinstance(g["table_alias"], str)
            lines[group].append((g["table_alias"], get_line(runs, grouping_keys, g["group"])))
    return lines


def generate_latex_table(groups: defaultdict[str, list[tuple[str, RunSummary]]]) -> str:
    table_lines = []

    table_lines.append(r"\begin{table}[H]")
    table_lines.append(r"    \centering")
    table_lines.append(r"    \begin{tabular}{lcccc}")
    table_lines.append(r"    \toprule")
    table_lines.append(r"    & \multicolumn{4}{c}{SimCLR - KNN5 Cross Video Accuracy} \\")
    table_lines.append(r"    \cmidrule(lr){2-5}")
    table_lines.append(r"    & Val & Test & All & Val-KFold \\")
    table_lines.append(r"    \midrule")

    for group_name, group_data in groups.items():
        table_lines.append(f"    \\multicolumn{{5}}{{l}}{{\\textbf{{{group_name}}}}} \\\\")

        for alias, run in group_data:
            metrics = run.metrics
            if any(map(lambda x: x != x, metrics.values())):
                table_lines.append(
                    f"    \\verb|{alias}| & - & - & - & - \\\\"
                )
                continue

            table_lines.append(
                f"    \\verb|{alias}| & {metrics['knn5-cv-val'] * 100:.1f}\\% & {metrics['knn5-cv-test'] * 100:.1f}\\% & {metrics['knn5-cv-all'] * 100:.1f}\\% & {metrics['knn5-kfold'] * 100:.1f}\\% \\\\"
            )

        table_lines.append(r"    \midrule")

    table_lines.append(r"    \bottomrule")
    table_lines.append(r"    \end{tabular}")
    table_lines.append(r"    \caption{Performance comparison of SimCLR across different parameters}")
    table_lines.append(r"    \label{tab:simclr-comparison}")
    table_lines.append(r"\end{table}")

    return "\n".join(table_lines)


runs = get_runs("Embedding-ViTLarge-SSL-Height")
groups = get_groups(runs, face_height)
latex_code = generate_latex_table(groups)
print(latex_code)
```
